#### Question 1

```python
query = 'select distinct(last_name) "last_name" from actor'
data = pd.read_sql_query(query, engine)
data.head()
```

#### Question 2

```python
query = ''' SELECT *,
            case \
            when weekday(rental_date)<6 then 'workday' \
            else 'weekend' \
            end as day_type  \
            FROM rental'''

data = pd.read_sql_query(query, engine)
data.head()
```

#### Question 3

```python
query = 'select * from film where title LIKE "%%ARMAGEDDON%%"'
data = pd.read_sql_query(query, engine)
data.head()
```

#### Question 4

```python
query = 'select * from film order by length desc limit 10'
data = pd.read_sql_query(query, engine)
data.head()
```

#### Question 5

```python
query = 'select count(title) as "Movies with behind the scenes" \
        from film \
        where special_features LIKE "%%Behind the Scenes%%"'
data = pd.read_sql_query(query, engine)
data.head()
```

#### Question 6

```python
query = 'select rating, round(avg(length),2) as mean \
        from film \
        group by rating \
        having mean>120'
data = pd.read_sql_query(query, engine)
data.head()
```

#### Question 7

```python
query = 'select title, length from film where length > 0 order by length desc'
data = pd.read_sql_query(query, engine)
data.head()
```
